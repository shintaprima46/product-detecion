```python
#read data
data = pd.read_csv("D:/SHINTA/workshop shopee code league 2020/shopee-product-detection-open/train.csv")
data.head()
```

```python
X = data.iloc[:,0:13] #Save first 13 column and all row to X
X.head()
```

```python
y = data['category']
y = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

```python
model = Sequential()
n_cols = X_train.shape[1] #GET NUMBER OF COLUMN

#add model layers
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(10, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
#Compiling Model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```
